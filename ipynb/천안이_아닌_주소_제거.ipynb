```python
import pandas as pd
```

```python
bus_location = pd.read_csv('/content/2022년_전국버스정류장 위치정보_데이터.csv')
```

```python
bus_cheon = bus_location[bus_location['도시명'] == '천안시']
bus_cheon
```

```python
bus_up_down = bus_cheon[['위도','경도']]
```

```python
import csv
from geopy.geocoders import Nominatim

# 데이터프레임에서 위도와 경도 열을 선택 (실제 데이터프레임이어야 합니다.)
bus_up_down = bus_cheon[['위도', '경도']]

# 가입 없이 좌표->주소 변환 함수 정의
def geocoding_reverse(lat_lng_str):
    geolocator = Nominatim(user_agent='YourUniqueUserAgent', timeout=None)
    location = geolocator.reverse(lat_lng_str, exactly_one=True)
    if location:
        return location.address
    else:
        return "주소를 찾을 수 없음"

# CSV 파일 경로 지정
output_csv_file = '주소_출력.csv'

# CSV 파일 열기 및 헤더 작성
with open(output_csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['위도', '경도', '주소']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # 각 행에 대한 주소 변환 및 CSV 파일에 쓰기
    bus_cnt = len(bus_up_down)
    for i in range(bus_cnt):
        lat = bus_up_down.iloc[i]['위도']
        lon = bus_up_down.iloc[i]['경도']
        lat_lng_str = f'{lat}, {lon}'
        address = geocoding_reverse(lat_lng_str)

        # CSV 파일에 쓰기
        writer.writerow({'위도': lat, '경도': lon, '주소': address})
```

```python
address = pd.read_csv('/content/주소_출력.csv')
address
```

```python
new_address = address[~address['주소'].str.contains('평택')]
new_address = new_address[~new_address['주소'].str.contains('안성')]
new_address = new_address[~new_address['주소'].str.contains('진천')]
new_address.to_csv('new_address.csv')
```

```python
merged_df = pd.merge(bus_cheon, new_address, on=['위도','경도'], how = 'inner')
merged_df.to_csv('only_cheonan.csv')
```
