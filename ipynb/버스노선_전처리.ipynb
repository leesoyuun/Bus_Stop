```python
# bus_stop 데이터프레임 읽기
bus_stop = pd.read_csv('./output_2.csv')

# 새로운 데이터프레임을 담을 딕셔너리 생성
bus_dataframes = {}

# for 루프를 사용하여 데이터프레임 생성 및 처리
for i in range(len(bus_stop)):
    # 각 행을 데이터프레임으로 생성
    bus_dataframes[i] = bus_stop.loc[[i]].melt()

# 새로운 데이터프레임을 만들기 위한 빈 리스트 생성
new_data = []

# 각 데이터프레임을 for 루프로 순회하며 데이터 처리
for i in range(len(bus_stop)):
    for j, row in enumerate(bus_dataframes[i].iloc[1::2].iterrows(), start=1):
        _, data = row
        route_name = data[0]
        direction = data[1]
        if pd.notna(direction):  # NaN 값이 아닌 경우에만 추가
            new_data.append([i + 1, j, direction])

# 추출한 데이터로 새로운 데이터프레임 생성
df_new = pd.DataFrame(new_data, columns=['버스노선번호', '몇번째 정류장', '노선이름'])

df_new.to_csv('output_busstop.csv',encoding='utf-8')
```
